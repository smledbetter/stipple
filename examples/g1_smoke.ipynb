# Stipple G1 smoke test

1M random points round-tripped through Arrow IPC over the comm channel.

Expected widget status: `✓ Stipple G1 — received 1,000,000 rows from Python` with Arrow IPC size and decode time.

```python
import numpy as np
from stipple import Stipple
import stipple
print(f"stipple {stipple.__version__}")
```

```python
rng = np.random.default_rng(0)
n = 1_000_000
x = rng.standard_normal(n).astype(np.float32)
y = rng.standard_normal(n).astype(np.float32)
print(f"prepared {n:,} points — {x.nbytes + y.nbytes:,} bytes raw")
```

```python
w = Stipple(x=x, y=y)
w
```

```python
import time
# Poll briefly so the async receive has a chance to update.
for _ in range(40):
    if w.rows_received:
        break
    time.sleep(0.1)
print(f"status         : {w.status}")
print(f"rows_received  : {w.rows_received:,}")
print(f"bytes_received : {w.bytes_received:,}")
```
