# P5.0 smoke — RasterScan density rendering

10M synthetic points across 5 Gaussian clusters. Renders in density mode: a
1024² per-bin count grid log-color-mapped through the viridis palette. The
render kernel does *not* touch the points buffer per-frame — cost scales
with rendered pixels, not with N.

```python
import numpy as np
from stipple import Stipple
import stipple
print(f'stipple {stipple.__version__}')
```

```python
rng = np.random.default_rng(11)
n = 10_000_000
centers = np.array(
    [[-3.0, 0.0], [0.0, 2.5], [3.0, 0.0], [-1.5, -2.0], [1.5, -2.0]],
    dtype=np.float32,
)
code = rng.integers(0, 5, size=n)
xs = centers[code, 0] + rng.standard_normal(n).astype(np.float32) * 0.7
ys = centers[code, 1] + rng.standard_normal(n).astype(np.float32) * 0.7
print(f'{n:,} points · 5 clusters')
```

```python
w = Stipple(x=xs, y=ys, render_mode='density')
w
```

```python
import time
for _ in range(60):
    if w.last_fps:
        break
    time.sleep(0.1)
print(f'status      : {w.status}')
print(f'render_mode : {w.render_mode}')
print(f'rows        : {w.rows_received:,}')
print(f'FPS         : {w.last_fps:.1f}')
print(f'frame_ms    : {w.avg_frame_ms:.2f}')
```
