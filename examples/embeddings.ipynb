# Embedding visualization with Stipple

Render an embedding projection, lasso a region of interest, get the row indices back in Python — the canonical ML-researcher workflow.

This example uses the `sklearn` digits dataset (1797 samples, 8×8 grayscale digits 0–9). PCA → 2D for display. Replace with your own embedding + UMAP / t-SNE / PCA in real use.

## Workflow
1. Compute (or load) a 2-D embedding
2. Pass `x`, `y`, `color` to `Stipple(...)`
3. Shift+drag in the canvas to lasso a region
4. Read `w.selected_indices` back in Python
5. Do something with those rows: inspect, retrain on, etc.

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from stipple import Stipple
```

```python
digits = load_digits()
X, y = digits.data, digits.target
proj = PCA(n_components=2).fit_transform(X)
labels = np.array([str(d) for d in y])
print(f'{X.shape[0]:,} samples · {X.shape[1]} features → 2D projection')
```

```python
w = Stipple(x=proj[:, 0], y=proj[:, 1], color=labels)
w
```

```python
  print("len:", len(w.selected_indices), "count:", w.selection_count, "id:", id(w))
```

Shift+drag in the canvas above to lasso a cluster of digits. Then run the next cell to inspect what you selected.

```python
import collections
idx = w.selected_indices
if len(idx) == 0:
    print('No selection yet. Shift+drag in the widget above to lasso a region, then re-run this cell.')
else:
    counts = collections.Counter(y[idx].tolist())
    print(f'Selected {len(idx):,} samples')
    print('Digit distribution:')
    for digit, c in sorted(counts.items()):
        bar = '█' * int(40 * c / max(counts.values()))
        print(f'  {digit}: {c:>4}  {bar}')
```
