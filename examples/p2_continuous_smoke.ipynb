# P2 smoke — continuous color

100k points where `color` is a continuous loss-like float. Should render as a viridis gradient with the banner reporting the [vmin, vmax] range.

```python
import numpy as np
from stipple import Stipple
import stipple
print(f'stipple {stipple.__version__}')
```

```python
rng = np.random.default_rng(2)
n = 100_000
# Color gradient runs diagonally so the viridis spread is visible.
t = np.linspace(0.0, 1.0, n, dtype=np.float32)
x = (t * 6.0 - 3.0) + rng.standard_normal(n).astype(np.float32) * 0.2
y = (t * 6.0 - 3.0) + rng.standard_normal(n).astype(np.float32) * 0.2
loss = t.astype(np.float32)
print(f'{n:,} points, loss range [{loss.min():.3f}, {loss.max():.3f}]')
```

```python
w = Stipple(x=x, y=y, color=loss, color_kind='continuous')
w
```

```python
import time
for _ in range(40):
    if w.last_fps:
        break
    time.sleep(0.1)
print(f'status        : {w.status}')
print(f'rows          : {w.rows_received:,}')
print(f'color_kind    : continuous (forced)')
print(f'color_range   : {w.color_range}')
print(f'categories    : {w.color_categories}  (should be empty)')
print(f'FPS           : {w.last_fps:.1f}')
```
