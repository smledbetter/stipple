# P1 smoke — DataFrame-positional API

Renders 100k points from a `pandas.DataFrame` using the column-name API:
`Stipple(df, x='col_x', y='col_y', color='label')`.

```python
import numpy as np
import pandas as pd
from stipple import Stipple
import stipple
print(f'stipple {stipple.__version__}')
```

```python
rng = np.random.default_rng(7)
n = 100_000
centers = np.array([[-3.0, 0.0], [0.0, 2.5], [3.0, 0.0]], dtype=np.float32)
labels = rng.choice(['cluster_a', 'cluster_b', 'cluster_c'], size=n)
code = np.searchsorted(['cluster_a', 'cluster_b', 'cluster_c'], labels)
xs = centers[code, 0] + rng.standard_normal(n).astype(np.float32) * 0.6
ys = centers[code, 1] + rng.standard_normal(n).astype(np.float32) * 0.6
df = pd.DataFrame({'feat_u': xs, 'feat_v': ys, 'label': labels})
print(df.head())
print(f'{len(df):,} rows · {df["label"].nunique()} classes')
```

```python
w = Stipple(df, x='feat_u', y='feat_v', color='label')
w
```

```python
import time
for _ in range(40):
    if w.last_fps:
        break
    time.sleep(0.1)
print(f'status      : {w.status}')
print(f'rows        : {w.rows_received:,}')
print(f'categories  : {w.color_categories}')
print(f'FPS         : {w.last_fps:.1f}')
```
